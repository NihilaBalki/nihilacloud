```python
f = open("student.txt", "a")

# Check if the file is empty
if f.tell() == 0:
    # Add titles to the first line
    titles = "S.No,Date,Name,Percentage,Bus Fare,Academic Fees,Total Fees,Installments,Remaining Amount,Paid Fees,Placements"
    f.write(titles + "\n")

sno = 1
number = 1
count = 1

while number > 0:
    emi = 0
    a = ""
    dept = ""
    percent = 0
    topay = 0

    print("4 departments:\n1.CSE\n2.ECE\n3.Civil\n4.Mech")
    something = input("Which department do you choose?: ")
    dept = something.upper()
    details(dept)

    ad = input("\nWould you like to proceed with admission? ")
    if ad.lower() in ["yes", "y"]:
        name = input("\nEnter your name: ")
        date1 = datetime.datetime.now()
        date2 = date1.strftime("%d-%m-%Y")
        datefinal = date2

        f.write(f"{sno},{datefinal},{name},")

        admn()

        a = input("\nWould you like to pay the fees now? ")
        percent = int(input("\nEnter your 12th percentage: "))
        f.write(f"{percent},")

        bus = input("Would you like to arrange for college transport? ")
        if bus.lower() in ["yes", "y"]:
            km = int(input("How far in kms are you away from college? "))
            busfare = transport(km)
            f.write(f"{km} = {busfare},")
        else:
            busfare = 0
            f.write("0 = 0,")

        topay = pay(a, dept, percent, count)
        z = int(topay)
        print("\nAcademic fees: Rs.", topay)
        f.write(f"{topay},")
        print("\nTotal fees: Rs.", topay + busfare)
        f.write(f"{topay + busfare},0,")

        ynemi = int(input("Would you like to pay by, Press 1 or 2 for:\n1.installments\n2.single payment?\n"))
        if z > 10000:
            if ynemi == 1:
                emi = yemi(z)
                f.write(f"0,{emi},")
                print(f"You have to pay Rs.{emi}/semester")

        p = input("\nAre you opting for placements? ")
        placement(p)

        if a.lower() in ["no", "n"]:
            # Skip asking the "Have you paid the fees?" question
            amt = 0
            k = paid(z, busfare, amt, emi)
            if ynemi == 1:
                print(f"Remaining amount this semester: {k}")
                f.write(f"0,{k},")
            else:
                print(f"Remaining amount this year: {k}")
                f.write(f"{k},0,")
        else:
            d = input("\nHave you paid the fees? ")
            if d.lower() in ["yes", "y"]:
                amt = int(input("Enter the amount you paid: "))
                f.write(f"{amt},")
                k = paid(z, busfare, amt, emi)
                if ynemi == 1:
                    print(f"Remaining amount this semester: {k}")
                    f.write(f"0,{k},")
                else:
                    print(f"Remaining amount this year: {k}")
                    f.write(f"{k},0,")

        f.write(f"{p}\n")
    else:
        f.write("0,0,0\n")
        bye()

    another = input("\n\n\nWould you like to enter another admission? ")
    if another.lower() in ["yes", "y"]:
        number = 1
        print("\n")
        count += 1
        sno += 1
    else:
        number = 0

print("\n\n")
last = "-----------Admission process end for today------------"
print(last.center(80))
f.close()
```

```python
import csv

class Student:
    def __init__(self, sno, date, name, percentage, bus_fare, academic_fees,total_fees, installments, remaining_amount, paid_fees, placements):
        self.sno = sno
        self.date = date
        self.name = name
        self.percentage = percentage
        self.bus_fare = bus_fare
        self.academic_fees = academic_fees
        self.total_fees = total_fees
        self.installments = installments
        self.remaining_amount = remaining_amount
        self.paid_fees = paid_fees
        self.placements = placements

    def __str__(self):
        return f"Roll No: {self.sno}\nDate: {self.date}\nName: {self.name}\nPercentage: {self.percentage}\nBus Fare: {self.bus_fare}\nAcademic Fees: {self.academic_fees}\nTotal Fees: {self.total_fees}\nInstallments: {self.installments}\nRemaining Amount: {self.remaining_amount}\nPaid Fees: {self.paid_fees}\nOpting for placements: {self.placements}"

def read_students_from_file():
    students = []
    with open("student.txt", "r") as file:
        reader = csv.reader(file)
        header = next(reader)  
        for row in reader:
            sno = row[0]
            date = row[1]
            name = row[2]
            percentage = row[3]
            bus_fare = row[4]
            academic_fees = row[5]
            total_fees = row[6]
            installments = row[7]
            remaining_amount = row[8]
            paid_fees = row[9]
            placements = row[10]
            student = Student(sno, date, name, percentage, bus_fare, academic_fees,total_fees, installments, remaining_amount, paid_fees, placements)
            students.append(student)
    return students

def display_student_details(student):
    if student:
        print(student)
    else:
        print("No student found.")

students = read_students_from_file()
choice = input("Enter 'name' or 'serial number' to search for a student: ")

if choice.lower() == "name":
    name = input("Enter student name: ")
    found_students = [student for student in students if student.name == name]
    for student in found_students:
        display_student_details(student)
elif choice.lower() == "serial number":
    sno = input("Enter student serial number: ")
    student = next((student for student in students if student.sno == sno), None)
    display_student_details(student)
else:
    print("Invalid choice. Please try again.")
```
